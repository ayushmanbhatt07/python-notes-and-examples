# File Handaling
## Type of file
1. ASCII Text Files(Text Stream)
    - A text file is a stream of characters that can be sequentially processed by a computer in forward direction.
    - For this reason a text file is usually opened for only one kind of operation (reading, writing, or appending)
    - in a text file each line of data ends with a newline character. Each file ends with a specail character called the end-of-file (EOF) marker
2. Binary Files(Byte Stream)
    - A binay file is which may contain  any type of data,encoded in binary form for computer storage and processing purpose
    - It includes files such as word processing documents,PDF, images, spreadsheets,video,zip file and other executable program
    - Binary file is collection of bytes
    - Binary file process sequentially or  randomly depending on the needs of the application

```python
# Example 1: Writing to a file
# Open a file in write mode ('w')
with open("example.txt", "w") as file:
    file.write("Hello, World!\n")
    file.write("This is a file handling example in Python.\n")
```

```python
# Example 2: Reading from a file
# Open a file in read mode ('r')
with open("example.txt", "r") as file:
    content = file.read()  # Read the entire file content
    print("File Content:\n", content)
```

```python
# Example 2: Reading from a file
# Open a file in read mode ('r')
with open("example.txt", "r") as file:
    content = file.read()  # Read the entire file content
    count = 0
    for i in content:
        if i in "aeiouAEIOU":
            count +=1
    print("No of vowels are", count)

```

```python
# Example 3: Appending to a file
# Open a file in append mode ('a')
with open("example.txt", "a") as file:
    file.write("This line is appended to the file.\n")
```

```python
# Example 4: Reading line by line
with open("example.txt", "r") as file:
    print("\nReading line by line:")
    for line in file:
        print(line.strip())  # strip() removes extra newline characters
```

```python
# Example 5: Using readlines()
with open("example.txt", "r") as file:
    lines = file.readlines()  # Returns a list of lines
    print("\nFile lines as a list:")
    print(lines)
```

```python
# List of lines to write
lines = [
    "This is line 1.\n",
    "This is line 2.\n",
    "This is line 3.\n"
]

# Open the file in write mode
with open("example.txt", "w") as file:
    file.writelines(lines)  # Write all lines to the file
```

```python
with open("example.txt", "r") as file:
    print("Initial position:", file.tell())  # Get current position (0)
    file.seek(10)  # Move the file pointer to the 10th byte
    print("Position after seek:", file.tell())
    content = file.read(5)  # Read 5 characters from the current position
    print("Content from position 10 to 15:", content)
```

```python
import csv

# Writing to a CSV file
data = [
    ["Name", "Age", "City"],
    ["Alice", 30, "New York"],
    ["Bob", 25, "Los Angeles"],
    ["Charlie", 35, "Chicago"]
]

with open("data.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)  # Write all rows to the CSV file

# Reading from a CSV file
with open("data.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
import json

# Writing to a JSON file
data = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}

with open("data.json", "w") as file:
    json.dump(data, file)  # Write JSON data to the file

# Reading from a JSON file
with open("data.json", "r") as file:
    loaded_data = json.load(file)  # Load JSON data from the file
    print("Loaded JSON data:", loaded_data)
```
