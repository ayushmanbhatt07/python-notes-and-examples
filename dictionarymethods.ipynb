dictionary can have sets tuples as elements

popitem function:pops up any random item from the dictionary.


```python
my_dict={'a':1,'b':2,'c':3}
print(my_dict.popitem())
```

deleting any item from dict:

```python
my_dict={'a':1,'b':2,'c':3}
del my_dict['b']
print(my_dict)
```

```python
my_dict={'a':1,'b':2,'c':3}
del my_dict
print(my_dict)
```

ADDING AND UPDATING ELEMENTAS IN THE DICTIONARY:

```python
my_dict={'a':1,'b':2,'c':3}
my_dict.update({'d':4})
print(my_dict)
```
